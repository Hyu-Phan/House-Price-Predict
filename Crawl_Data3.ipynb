```python
import undetected_chromedriver as uc
from selenium import webdriver
from bs4 import BeautifulSoup

options = webdriver.ChromeOptions()
options.add_argument("start-maximized")
options.add_argument("--incognito")
driver = uc.Chrome(options=options)
```

```python
links = []
```

```python
links = []
for i in range(1,201):
    url = f'https://batdongsan.com.vn/ban-nha-rieng-ha-noi/p{i}'
    driver.get(url)
    soup = BeautifulSoup(driver.page_source)
    for link in soup.find_all('a', attrs={'class': 'js__product-link-for-product-id'}):
        links.append(link['href'])
len(links)
```

```python
links = []
for i in range(300, 330):
    url = f'https://batdongsan.com.vn/ban-nha-rieng-ha-noi/p{i}?gtn=0-trieu&gcn=10-ty'
    driver.get(url)
    soup = BeautifulSoup(driver.page_source)
    for link in soup.find_all('a', attrs={'class': 'js__product-link-for-product-id'}):
        links.append(link['href'])
len(links)
```

```python
with open(r'links3.txt', 'w') as fp:
    for link in links:
        # write each item on a new line
        fp.write("%s\n" % link)
    print('Done')

```

```python
links =[]
with open(r'links3.txt', 'r') as fp:
    for line in fp:
        # print(line)
        links.append(line[:-1])
```

```python

import pandas as pd
list_data = []
i = 0
for link in links[:300]:
    try:
        i+=1
        print(i)
        driver.delete_all_cookies()
        driver.get('https://batdongsan.com.vn' + link)
        dic = {}
        soup = BeautifulSoup(driver.page_source)
        dic['Địa chỉ'] = soup.find(
            'span', attrs={'class': 're__pr-short-description js__pr-address'}).get_text()
        for div in soup.find_all('div', attrs={'class': 're__pr-specs-content-item'}):
            title = div.find(
                'span', attrs={'class': 're__pr-specs-content-item-title'}).get_text()
            value = div.find(
                'span', attrs={'class': 're__pr-specs-content-item-value'}).get_text()
            dic[title] = value
        list_data.append(dic) 
    except Exception as e:
        print(e)
```

```python
import pandas as pd
data = pd.DataFrame(list_data)
data.to_csv('test.csv', index=False)
```
